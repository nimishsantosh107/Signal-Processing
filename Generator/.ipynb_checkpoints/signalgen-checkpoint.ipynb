```python
import wave, struct, random, math, heapq
import pyaudio
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
from scipy.fftpack import fft,fftfreq
%matplotlib inline
```

## Functions

```python
def stringToBits(st):
    result = []
    for char in st:
        bits = bin(ord(char))[2:]
        bits = '00000000'[len(bits):] + bits
        for bit in bits:
            result.append(int(bit))
    return result

def bitsToString(bits):
    chars = []
    for b in range(len(bits) // 8):
        byte = bits[b*8:(b+1)*8]
        chars.append(chr(int(''.join([str(bit) for bit in byte]), 2)))
    return ''.join(chars)
```

```python
def genSignal(bits, outfile, n_channels, sample_width, sampling_rate, n_samples, amplitude, F1, F2):
    wav_file=wave.open(outfile, 'w')
    wav_file.setnchannels(n_channels)
    wav_file.setsampwidth(sample_width)
    wav_file.setframerate(sampling_rate)
    wav_file.setnframes(n_samples)
    for i in range(len(bits)):
        SEGMENT = n_samples // len(bits)
        F = F1 if(bits[i] == 0) else F2
        sine_wave = [np.sin(2 * np.pi * F * x/(sampling_rate)) for x in range( (i*SEGMENT) , ((i+1)*SEGMENT) )]
        for s in sine_wave:
           wav_file.writeframesraw(struct.pack('<h', int(s * amplitude)))
    wav_file.close()
```

## Main

```python
#General
FREQUENCY1 = 20000
FREQUENCY2 = 19000
N_CHANNELS = 1
SAMPLING_RATE = 48000 #Hz
SAMPLE_WIDTH = 2 #bytes
AMPLITUDE = 2000 #0-32767

#file specific
FILE = "./../sound.wav"
STRING = "TESTING"
N_SAMPLES = SAMPLING_RATE * len(STRING)
```

```python
BITS = stringToBits(STRING)
genSignal(BITS, FILE, N_CHANNELS, SAMPLE_WIDTH, SAMPLING_RATE, N_SAMPLES, AMPLITUDE, FREQUENCY1, FREQUENCY2)
```
