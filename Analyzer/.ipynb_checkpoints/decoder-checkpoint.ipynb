```python
import wave, struct, random, math, heapq
import pyaudio
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
from scipy.fftpack import fft,fftfreq
%matplotlib inline
```

```python
def stringToBits(st):
    result = []
    for char in st:
        bits = bin(ord(char))[2:]
        bits = '00000000'[len(bits):] + bits
        for bit in bits:
            result.append(int(bit))
    return result

def bitsToString(bits):
    chars = []
    for b in range(len(bits) // 8):
        byte = bits[b*8:(b+1)*8]
        chars.append(chr(int(''.join([str(bit) for bit in byte]), 2)))
    return ''.join(chars)
```

```python
#General
FREQUENCY1 = 20000
FREQUENCY2 = 19000
N_CHANNELS = 1
SAMPLING_RATE = 48000 #Hz
SAMPLE_WIDTH = 2 #bytes
AMPLITUDE = 2000 #0-32767

#file specific
FILE = "./../sound.wav"
```

```python
ipd.Audio(FILE)
```

```python
wav_file = wave.open(FILE, 'r')
N_SAMPLES = wav_file.getnframes()
audio = wav_file.readframes(N_SAMPLES)
wav_file.close()

audio = struct.unpack('{n}h'.format(n=N_SAMPLES), audio)
audio = np.array(audio)
```

```python
SEGMENTS = SAMPLING_RATE // 8
result = []
for i in range(0,N_SAMPLES//SEGMENTS):
    section = audio[(i*SEGMENTS) : ((i+1)*SEGMENTS)]
    section_fft = fft(section)
    frequencies = np.abs(section_fft)[0:SEGMENTS//2]
    max_peaks = heapq.nlargest(1, frequencies)
    result.append(np.where(frequencies == max_peaks[0])[0][0])
```

```python
maxf = max(set(result))
print(set(result))
binresult = []
for i in result:
    if(i == maxf):
        binresult.append(int(0))
    else:
        binresult.append(int(1))
```
